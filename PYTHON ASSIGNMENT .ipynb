Question 1: -
Write a program that takes a string as input, and counts the frequency of each word in the string, there might
be repeated characters in the string. Your task is to find the highest frequency and returns the length of the
highest-frequency word.

Note - You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.
Example input - string = “write write write all the number from from from 1 to 100”
Example output - 5
Explanation - From the given string we can note that the most frequent words are “write” and “from” and
the maximum value of both the values is “write” and its corresponding length is 5

```python
def find_highest_frequency_word_length(string):
    # Split the string into words
    words = string.split()

    # Create a dictionary to store the frequency of each word
    frequency_dict = {}
    for word in words:
        frequency_dict[word] = frequency_dict.get(word, 0) + 1

    # Find the word with the highest frequency
    max_frequency = max(frequency_dict.values())
    highest_frequency_word = max(word for word, freq in frequency_dict.items() if freq == max_frequency)

    # Return the length of the highest-frequency word
    return len(highest_frequency_word)


# Test case 1
string1 = "write write write all the number from from from 1 to 100"
print(find_highest_frequency_word_length(string1))  # Output: 5

# Test case 2
string2 = "the quick brown fox jumps over the lazy dog"
print(find_highest_frequency_word_length(string2))  # Output: 3

# Test case 3
string3 = "hello hello world world world"
print(find_highest_frequency_word_length(string3))  # Output: 5
```

Consider a string to be valid if all characters of the string appear the same number of times. It is also valid if
he can remove just one character at the index in the string, and the remaining characters will occur the same
number of times. Given a string, determine if it is valid. If so, return YES , otherwise return NO .

Note - You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.
Example input 1 - s = “abc”. This is a valid string because frequencies are { “a”: 1, “b”: 1, “c”: 1 }
Example output 1- YES
Example input 2 - s “abcc”. This string is not valid as we can remove only 1 occurrence of “c”. That leaves
character frequencies of { “a”: 1, “b”: 1 , “c”: 2 }
Example output 2 - NO

```python
def is_valid_string(s):
    # Count the frequency of each character in the string
    char_count = {}
    for char in s:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    
    # Count the frequency of frequencies
    freq_count = {}
    for count in char_count.values():
        if count in freq_count:
            freq_count[count] += 1
        else:
            freq_count[count] = 1
    
    # If there is only one distinct frequency, or if there are two distinct frequencies
    # and one of them occurs only once, the string is valid
    if len(freq_count) == 1 or (len(freq_count) == 2 and (1 in freq_count.values())):
        return "YES"
    else:
        return "NO"
```

```python
print(is_valid_string("abc"))  # Output: YES
print(is_valid_string("abcc"))  # Output: NO
```

```python
#Testing a valid string with multiple characters occurring the same number of times:
print(is_valid_string("aabbbcc"))  # Output: YES
#Explanation: In this case, the character frequencies are { "a": 2, "b": 3, "c": 2 }. All characters occur the same number of times, so the string is valid.
```

```python
#Testing an invalid string where more than one character needs to be removed:
print(is_valid_string("aabbccdd"))  # Output: NO
#Explanation: In this case, the character frequencies are { "a": 2, "b": 2, "c": 2, "d": 2 }. If we remove any single character, there will still be two characters remaining with a frequency of 2, but the other two characters will have a frequency of 1. Thus, it is not possible to make all characters occur the same number of times by removing just one character.
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from collections import defaultdict

def count_pos(text):
    # Tokenize the text into words
    words = word_tokenize(text)

    # Perform Part-of-Speech tagging
    tagged_words = pos_tag(words)

    # Create a defaultdict to store the counts
    pos_counts = defaultdict(int)

    # Count the occurrences of each POS tag
    for _, tag in tagged_words:
        if tag.startswith('VB'):
            pos_counts['Verbs'] += 1
        elif tag.startswith('NN'):
            pos_counts['Nouns'] += 1
        elif tag.startswith('PRP') or tag.startswith('WP'):
            pos_counts['Pronouns'] += 1
        elif tag.startswith('JJ'):
            pos_counts['Adjectives'] += 1

    return pos_counts

# Example usage
text = "The quick brown fox jumps over the lazy dog."
counts = count_pos(text)
print(counts)
```

```python
import nltk
from nltk import pos_tag
from nltk.tokenize import word_tokenize

def count_pos_tags(text):
    # Tokenize the text into words
    words = word_tokenize(text)

    # Perform part-of-speech tagging
    tagged_words = pos_tag(words)

    # Initialize count variables
    count_verbs = 0
    count_nouns = 0
    count_pronouns = 0
    count_adjectives = 0

    # Count the occurrences of each part of speech
    for word, tag in tagged_words:
        if tag.startswith('VB'):
            count_verbs += 1
        elif tag.startswith('NN'):
            count_nouns += 1
        elif tag.startswith('PRP'):
            count_pronouns += 1
        elif tag.startswith('JJ'):
            count_adjectives += 1

    # Create a dictionary with the count of each part of speech
    pos_counts = {
        'Verbs': count_verbs,
        'Nouns': count_nouns,
        'Pronouns': count_pronouns,
        'Adjectives': count_adjectives
    }

    return pos_counts

# Example usage
phrase = "The cat jumped over the lazy dog."
counts = count_pos_tags(phrase)
print(counts)
```
